<font color='red'>아래 코드는 내 구글 드라이브에 저장된 사진을 조작하기 위해 코랩으로 불러옵니다

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
img = io.imread('/content/drive/MyDrive/AI/IMG_8219.jpg')
io.imshow(img)
```

<font color='red'>아래 코드는 나의 privacy(개인정보, 사생활)를 보호하기 위해 이미지를 crop하여 내 얼굴을 잘라내기 위한 코드입니다. 이미지의 좌표를 설정하여 수행할 수 있습니다.

```python
crop_img = img[1000:3000, 0:2000]
io.imshow(crop_img)
```

<font color='red'>아래 코드는 다른 사용자가 사진을 통해 내 얼굴을 보는 것을 방해하기 위해 사진을 거꾸로 회전시킵니다. import한 모듈의 rotate 메소드를 사용하여 수행할 수 있습니다.

```python
from skimage.transform import rotate
img_rotate = rotate(img, 180, resize=True)
fig, ax = plt.subplots()
ax.imshow(img_rotate)
ax.set_title('image rotate') #이미지 이름 설정
```

<font color='red'>아래 코드는 얼굴 보호를 위해 색상 코드를 정의하여 지정하고, 그림(선)을 그릴 좌표와 범위를 변수 BB에 저장하여 얼굴의 주요 부위인 눈을 가리는 코드입니다.

```python
from skimage import draw
R = np.array([255,0,0],dtype=np.uint8) #빨간색 색 코드
BB = (300, 800, 1400, 400) # Bounding Box format (x, y, 너비, 높이)
img[BB[1]:BB[1]+BB[3], BB[0]:BB[0]+BB[2]] = R
io.imshow(img)
```
